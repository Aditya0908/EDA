## GROUP BY

the last column that gives the values dont have a name so to do we'll use firstly reset_index() option to assign a integer value to the column and then rename the same

```python
final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index()
```

```python
ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
ratings
```

jaha par bhi column ka naam 0 tha rename it to rating count

This code performs a grouping and aggregation operation on a DataFrame called `final_df`. Here's a breakdown of what each part of the code does:

1. `final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text'])`: This groups the DataFrame `final_df` by the columns 'Aggregate rating', 'Rating color', and 'Rating text'. It essentially creates groups based on unique combinations of values in these columns.

2. `.size()`: This calculates the size of each group, i.e., the number of rows in each group.

3. `.reset_index()`: This resets the index of the resulting DataFrame, so that the group keys become regular columns.

4. `.rename(columns={0:'Rating Count'})`: This renames the column resulting from the size calculation to 'Rating Count'.

The overall purpose of this code is to determine the count of rows (or occurrences) for each unique combination of 'Aggregate rating', 'Rating color', and 'Rating text' in the `final_df` DataFrame. The resulting DataFrame, `ratings`, will have columns 'Aggregate rating', 'Rating color', 'Rating text', and 'Rating Count', representing the unique combinations and their corresponding counts.

```python
sns.barplot(x='Aggregate rating', y= 'Rating Count', data=ratings)
```

to improve the quality of data being displayed

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x='Aggregate rating', y= 'Rating Count', data=ratings)
```
